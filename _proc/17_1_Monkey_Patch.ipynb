## ­¤¦Ā 1. What is Monkey Patching?

> **Monkey patching** is the practice of **dynamically changing a class, method, or module at runtime**, usually to alter or extend behavior without modifying the original source code.

Ō£ģ Often used in:

* Testing/mocking
* Temporary bug fixes
* Dynamic feature injection

## ŌÜĀ’ĖÅ 7. Risks of Monkey Patching

| Risk                               | Description                                              |
| ---------------------------------- | -------------------------------------------------------- |
| ­¤öä **Global side effects**         | Patches affect all usage of the object across the app    |
| ­¤ś▒ **Hard to trace**               | Dynamic behavior can confuse maintainers or IDEs         |
