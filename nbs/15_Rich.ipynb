## üñºÔ∏è What is Rich?

**`rich`** is a Python library for **rich text and beautiful formatting** in the terminal ‚Äî including:

* üåà Colored text & styles
* üìä Tables, progress bars, trees
* ü™µ Log rendering
* üì¶ JSON + syntax highlighting
* üñºÔ∏è Markdown and emoji rendering

It's used in projects like **Textual**, **FastAPI**, **Typer**, **pipx**, and **Pydantic**.

## üß± Basic Usage

```python
from rich import print

print("[bold magenta]Hello[/bold magenta] [green]World![/green] :rocket:")
```

‚úÖ Output with colors and emoji.

### Rich Console

```python
from rich.console import Console

console = Console()
console.print("Hello [cyan]World[/cyan]!", style="bold")
```

---

## üßæ Text Formatting

Supports:

* Styles: `bold`, `italic`, `underline`, `strike`
* Colors: names, hex (`#ff00ff`), RGB tuples
* Emoji shortcodes (like `:rocket:`)

```python
console.print("This is [bold red]error[/bold red] and [green]success[/green]!")
```

---

## üìê Layouts

### ‚úÖ Tables

```python
from rich.table import Table

table = Table(title="User Info")
table.add_column("Name", style="cyan")
table.add_column("Email", style="magenta")
table.add_row("Ben", "ben@example.com")
table.add_row("Kiera", "kiera@example.com")

console.print(table)
```

### üå≥ Trees

```python
from rich.tree import Tree

tree = Tree("üìÅ Root")
tree.add("üìÑ File1.txt")
branch = tree.add("üìÇ Folder")
branch.add("üìÑ NestedFile.txt")

console.print(tree)
```

### üîÑ Progress Bars

```python
from rich.progress import track
import time

for step in track(range(10), description="Processing..."):
    time.sleep(0.1)
```

### ‚è≥ Manual Progress Bars

```python
from rich.progress import Progress

with Progress() as progress:
    task = progress.add_task("Downloading...", total=100)
    while not progress.finished:
        progress.update(task, advance=10)
        time.sleep(0.2)
```

## üêû Logging with Rich

```python
import logging
from rich.logging import RichHandler

logging.basicConfig(
    level="DEBUG",
    handlers=[RichHandler()],
    format="%(message)s",
)

logger = logging.getLogger("rich")
logger.info("Info log")
logger.error("Error log")
```

---

## üß¨ Syntax Highlighting

```python
from rich.syntax import Syntax

code = '''
def greet(name):
    print(f"Hello, {name}")
'''

syntax = Syntax(code, "python", theme="monokai", line_numbers=True)
console.print(syntax)
```

---

## üìö Pretty Print JSON

```python
from rich.pretty import pprint
from rich.json import JSON

pprint({"foo": "bar", "baz": [1, 2, 3]})

console.print(JSON('{"key": "value", "list": [1, 2, 3]}'))
```

---

## ‚ú® Markdown Rendering

```python
from rich.markdown import Markdown

md = Markdown("# Title\n\n**Bold** and *italic* text.")
console.print(md)
```

---

## üí• Error Tracebacks

```python
from rich.traceback import install
install()

# Raise any exception after this and it'll show a beautiful traceback
raise ValueError("Oops!")
```

## üß™ Testing & Integration

### Django Logging

Use `RichHandler` in `settings.py`:

```python
LOGGING = {
    "version": 1,
    "handlers": {
        "console": {
            "class": "rich.logging.RichHandler",
            "level": "DEBUG",
        },
    },
    "root": {
        "handlers": ["console"],
        "level": "INFO",
    },
}
```

## üßµ Combining Components

```python
from rich.panel import Panel
from rich.align import Align

panel = Panel.fit(
    Align.center("[bold green]Task Complete![/bold green]"),
    border_style="green",
    title="Status",
)

console.print(panel)
```

---

## ‚öôÔ∏è Advanced: Live Dashboards

```python
from rich.live import Live
from rich.table import Table
import time

table = Table()
table.add_column("Task")
table.add_column("Progress")

with Live(table, refresh_per_second=4):
    for i in range(5):
        table.add_row(f"Task {i}", f"{i*20}%")
        time.sleep(0.5)
```
